```python
# Import required libraries
import pandas as pd
import numpy as np
from mlxtend.preprocessing import TransactionEncoder


df_cust_purchases = pd.read_csv('/content/drive/MyDrive/Market_Basket_Optimisation.csv')
df_cust_purchases.head()
```

```python
# Apply one hot encoding on toy dataset
toy_dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

        ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

        ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

        ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

      ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]

te = TransactionEncoder()
te_array = te.fit_transform(toy_dataset)


te_df = pd.DataFrame(te_array, columns = te.columns_)
te_df
```

```python
# Import and run apriori algorithm
from mlxtend.frequent_patterns import apriori
df_freq = apriori(te_df, min_support=0.6)
df_freq
```

```python
# frequent items
freq_items = apriori(te_df, min_support=0.6, use_colnames= True)
freq_items
```

```python
# Display number of freq elements
freq_items['lenght'] = df_freq['itemsets'].apply(lambda x: len(x))
freq_items
```

```python
# Import and run the association rules
from mlxtend.frequent_patterns import association_rules

rules_confidence = association_rules(freq_items, metric='confidence', min_threshold=0.7)
rules_confidence
```

```python
# Association rules with lift metric
rules_lift = association_rules(freq_items, metric= 'lift', min_threshold=0.75)
rules_lift
```

From the result obtained from the toy dataset, the frequently bought items are Skirts and Pants. There's a high probability of buying  Pants after buying Skirt as indicated by the Confidence metric(1). Also, lift value of 1.25 means that customers are more likely to buy skirts and pants together rather than buying only one of them.

```python
# Query dataset to find column contents
df_cust_purchases.query(" shrimp == ['shrimp','almonds','avocado','vegetables mix','green grapes','whole weat flour','yams'] ")
```

```python
# Reshaping dataset to have all items in a single column
x_transactions = []
for index, row in df_cust_purchases.iterrows():
  transaction = [str(item) for item in row if pd.notna(item)]
  x_transactions.append(transaction)

# Convert to dataframe
transaction_df = pd.DataFrame({'Items': x_transactions})
transaction_df.head()
```

```python
# Encode items data using transaction encoder
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
encoded_array = te.fit(transaction_df['Items']).transform(transaction_df['Items'])
encoded_df = pd.DataFrame(encoded_array, columns=te.columns_)
encoded_df
```

```python
# Applying apriori algorithm
from mlxtend.frequent_patterns import association_rules

freq_items = apriori(encoded_df, min_support=0.01, use_colnames=True)
freq_items.sort_values(by='support', ascending=False)
```

```python
# Generate association rules with confidence metric
rules_confidence = association_rules(freq_items, metric = 'confidence', min_threshold=0.4)
rules_confidence.sort_values(by='confidence', ascending=False)
```

```python
# Generate Association rules with Lift metric
rules_lift = association_rules(freq_items, metric='lift', min_threshold=1.0)
rules_lift.sort_values(by='lift', ascending=False).head(7)
```

## Result Interpretation and Business Plan Suggestion

###Frequent Items:
The frequently bought items from the Supermarket in descending order using a minimum support threshold of 0.08 are: Mineral water, Eggs, Spaghetti, French fries, Chocolate, Green tea, Milk, Ground beef, Frozen vegetables Pancakes, Burgers, Cakes, Cookies.

###Confidence Metric:
Customers who buy Eggs and Ground Beef or milk and Ground beef are likely to buy Mineral water after, with a probability of 0.5. The result also shows that customers who buy combination of other food items always buy Mineral water after.

###Lift Metric:
Customers mostly buy (Ground beef and herb & Pepper) together rather than buying them separately, same applies to (Mineral water & Spaghetti and Ground beef), (Olive oil and Mineral water & Spaghetti), (Tomatoes and Frozen vegetables)

###Business Plan Suggestion to Supermaket:
Always stock up the store with the frequently bought items.
Place Mineral water in a way that it can be easily found from where Ground beef and other food items are located and label appropriately.
Arrange the items customers prefer buying together in one place- (Ground beef and herb & Pepper), (Mineral water & Spaghetti and Ground beef) and possible offer a discounted rate when they buy them together.
